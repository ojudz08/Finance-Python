```python
# Matplotlib is a very useful library where you can make graphs from 
# given data inputs

# Example: Create a function to find the total cost based on a fixed
# cost and a variable cost
def totalCost(fixedCost, variableCost, volume):
    return fixedCost + variableCost * volume

xValues= list(range(0, 101))
yValues = [totalCost(50, 1, x) for x in xValues]
print(xValues)
print()
print(yValues)

import matplotlib.pyplot as plt
plt.plot(xValues, yValues)
plt.show()
```

```python
# xlabel(), ylabel() and title() --> Adding labels, titles 
plt.plot(xValues, yValues)
plt.xlabel("Units")
plt.ylabel("Total Cost")
plt.title("Total Cost")
plt.show()
```

```python
# Adding a second plot, revenue
def revenue(price, volume):
    return price * volume


xValues = list(range(0, 101))
yValues1 = [totalCost(50, 1, x) for x in xValues]
yValues2 = [revenue(2,x) for x in xValues]

plt.plot(xValues, yValues1)
plt.plot(xValues, yValues2)
plt.xlabel("Units")
plt.ylabel("Total Cost/Revenue")
plt.title("Revenue vs. Cost")
plt.show()
```

```python
# Graph legends
# plt.legend() 
plt.plot(xValues, yValues1)
plt.plot(xValues, yValues2)
plt.xlabel("Units")
plt.ylabel("Total Cost/Revenue")
plt.title("Revenue vs. Cost")
plt.legend(["Cost", "Revenue"])
plt.show()
```

```python
# Other solution is to give each line a label and call plt.legend() with no arguments
plt.plot(xValues, yValues1, label = "Cost")
plt.plot(xValues, yValues2, label = "Revenue")
plt.xlabel("Units")
plt.ylabel("Total Cost/Revenue")
plt.title("Revenue vs. Cost")
plt.legend()
plt.show()
```

```python
# Color plotting
plt.plot(xValues, yValues1, label="Cost", color="r")
plt.plot(xValues, yValues2, label="Revenue", color="g")

plt.xlabel("Units")
plt.ylabel("Total Cost/Revenue")
plt.title("Revenue vs. Cost")
plt.legend()
plt.show()
```

```python
# Find the profit by using zip to iterate over the total cost list and the revenue list
yValues3 = [a-b for a,b in zip(yValues1, yValues2)]
print(yValues3)

plt.plot(xValues, yValues1, label="Cost", color="r")
plt.plot(xValues, yValues2, label="Revenue", color="g")
plt.plot(xValues, yValues3, label="Profit", color="k")

plt.xlabel("Units")
plt.ylabel("Total Cost/Revenue")
plt.title("Reveneu vs. Cost")
plt.legend()
plt.show()
```

```python
print(xValues[::10])
print(yValues3[::10])

plt.plot(xValues, yValues1, label="Cost", color="r")
plt.plot(xValues, yValues2, label="Revenue", color="g")
plt.plot(xValues[::10], yValues3[::10], "ko", label="Profit")

plt.xlabel("Units")
plt.ylabel("Total Cost/Revenue")
plt.title("Revenue vs. Cost")
plt.legend()
plt.show()
```
